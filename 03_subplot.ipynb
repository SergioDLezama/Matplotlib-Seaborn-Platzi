```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos dos array para trabajar
x = np.linspace(0,5,11)
y = x ** 2
```

```python
'''
Subplot Syntax
.subplot(filas,numero de columnas,index que vamos a manejar entre filas y columnas)
Las caracteristicas que queramos poner no se afectan antre los plot
'''

# Definimos el primer plot, con index 1
plt.subplot(1,2,1)
plt.plot(x,y,'r:')

# Definimos el segundo plot, con index 2
plt.subplot(1,2,2)
plt.hist(y)

plt.show
```

```python
# Definimos el primer plot, con index 1
# Como solo definimos un index, pone ambas lineas en un solo plot
plt.subplot(1,2,1)
plt.plot(x,y,'r:')
plt.plot(y,x,'b:')

# Definimos el segundo plot, con index 2
plt.subplot(1,2,2)
plt.pie(y)

plt.show
```

```python
# Podemos jugar con el como ubicar las grficas
# Modificando el numero de filas y columnas podemos poner las graficas diferente
# Definimos el primer plot, con index 1
plt.subplot(2,1,1)
plt.plot(x,y,'r:')
plt.plot(y,x,'b:')

# Definimos el segundo plot, con index 2
plt.subplot(2,1,2)
plt.pie(y)

plt.show
```
