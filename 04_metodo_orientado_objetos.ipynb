```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# Para el MOO tenemos que crear un objeto
fig = plt.figure()

# Las figuras que vamos a crear son llamados axes
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')

fig.show()
```

```python
# Para el MOO tenemos que crear un objeto
fig = plt.figure()

# Figura mas delgada
axes = fig.add_axes([0.1,0.1,0.3,0.9])
axes.plot(x,y,'b')

fig.show()
```

```python
# Para el MOO tenemos que crear un objeto
fig = plt.figure()

# Figura mas larga
axes = fig.add_axes([0.1,0.1,0.5,1.2])
axes.plot(x,y,'b')

fig.show()
```

```python
# Para el MOO tenemos que crear un objeto
fig = plt.figure()
'''
axes Syntax
fig.add_axes([Pos. Eje X,Pos. Eje Y,Size graph eje X,Size graph eje Y])
'''

# Con los dos primeros valores de Syntax podemos cambiar el lugar del plot
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes_2.plot(y,x,'r')

fig.show()
```

```python
# Para el MOO tenemos que crear un objeto
fig = plt.figure()
'''
axes Syntax
fig.add_axes([Posicion figura en el lienzo,Posicion figura en el lienzo,Tamano de figura,Tamano de figura])
'''

# Con los dos primeros valores de Syntax podemos cambiar el lugar del plot
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes_2.plot(y,x,'r')

# Podemos cambiar el color del fondo con .facecolor('Color')
axes_2.set_facecolor('gray')

fig.show()
```
