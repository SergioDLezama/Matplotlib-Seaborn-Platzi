```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Definimos nuestra figura
fig, axes = plt.subplots()
```

```python
# Definimos nuestra figura
fig, axes = plt.subplots()

# Definimos nuestro axes
axes.plot(x,y,'b')
```

```python
# Para tener dos graficos en una figura
''' No siempre hay que definir nrows o ncols
El primer parametro siempre es filas, el segundo columnas'''
fig, axes = plt.subplots(nrows=1, ncols=2)

# Definimos nuestro axes
#axes.plot(x,y,'b')
```

```python
# Para tener dos graficos en una figura
fig, axes = plt.subplots(1, 2)

# Definimos nuestros axes mediante index
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
# Otro ejemplo
fig, axes = plt.subplots(2, 2)

axes[0,0].plot(x,y,'b')
axes[0,1].plot(y,x,'r')
axes[1,1].plot(x,y,'b')
axes[1,0].plot(y,x,'r')
```

```python
# Otra forma de definir los axes
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
# Otra forma de definirlo
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)

ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
```

```python
# Para que visualmente sea mejor usamos fig.tight_layout()
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)

ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)

fig.tight_layout()
```

```python
# Tarea
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)

# Fila 1
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)

# Fila 2

ax5.plot(y,np.cos(y))
ax6.plot(y,np.sin(y),'y')
ax7.plot(y,np.tan(y),'r')
ax8.plot(y,np.cos(y)**2)


fig.tight_layout()
```
