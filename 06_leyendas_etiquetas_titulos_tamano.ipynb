```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos variables
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Definimos la figura
fig, axes = plt.subplots(1,2)
```

```python
# Definimos la figura
fig, axes = plt.subplots(1,2)

# Definimos los axes
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# Definimos la figura
fig, axes = plt.subplots(1,2)

# Para poner titulo usamos axes.set_title()
axes[0].plot(x,y)
axes[0].set_title('Relacion X Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion Y X')
```

```python
fig, axes = plt.subplots(1,2)

# Para poner titulo usamos axes.set_title()
# Podemos poner un label a nuestros ejes con axes[n].set_xlabel() o axes[n].set_ylabel()
axes[0].plot(x,y)
axes[0].set_title('Relacion X Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion Y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
# Podemos modificar el tamano de la figura con figsize={nx,ny}
fig, axes = plt.subplots(1,2,figsize=(10,5))

axes[0].plot(x,y)
axes[0].set_title('Relacion X Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Relacion Y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))

'''
Para definir labels en leyenda usamos:
Dentro de el codigo del plot label='Texto'
axes[n].legend()
'''

axes[0].plot(x,y,label='sin(X)')
axes[0].set_title('Relacion X Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
'''En caso de no querer usar el metodo oriendado a objetos'''
plt.plot(x,y,label='sin(X)')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')

# Podemos cambiar la ubicacion de la leyenda con loc='location'
plt.legend(loc='lower left')
```

```python
'''En caso de no querer usar el metodo oriendado a objetos'''
plt.plot(x,y,label='sin(X)')
plt.title('Este es un titulo')
plt.xlabel('X')
plt.ylabel('Y')

# Podemos nosotros definir la ubicacion con bbox_to_anchor=(nx,ny)
plt.legend(loc='lower left',bbox_to_anchor=(0.7,0.3))
```
