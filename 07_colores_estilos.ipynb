```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos variables
x = np.linspace(0,5,11)
```

```python
# Asi vemos los estilos que tenemos de default
print(plt.style.available)
```

```python
# Definimos el estilo que vamos a usar
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r')
ax.plot(x,x+2,'orange')
ax.plot(x,x+3,'y')
ax.plot(x,x+4,'g')
```

```python
''' 
Modificaciones estilo Matlab
'''
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'orange')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

```python
''' 
Modificaciones estilo Pyplot
alpha= n Nos dara transparencia en la linea
linewidth= n Modifica lo grueso de la linea
linestyle= 'stilo' Cambia el estilo de la linea
marker= 'tipo' Define el marcador
markersize= 'n' Modifica el tamano de el marker
markerfacecolor= 'color' Modifica el color del marker
'''
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color= 'green', alpha=0.5, linewidth=5, linestyle='-', marker='x', 
markersize=9, markerfacecolor='white')
ax.plot(x,x+2,color= 'orange', linewidth=8, linestyle='--', marker='o', markersize=5)
ax.plot(x,x+3,color= '#66ff66', linewidth=3, linestyle='dashed', marker='v',
markersize=15, markerfacecolor='yellow')
ax.plot(x,x+4,color= 'white', linewidth=2, linestyle=':', marker='P')
```
