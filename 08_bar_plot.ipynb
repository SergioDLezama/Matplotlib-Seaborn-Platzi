```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definimos el estilo que vamos a usar
plt.style.use('dark_background')
```

```python
# Definimos variables categoricas

country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1100,300]
```

```python
'''
Junto a las variables categoricas que creamos haremos una grafica de barras 
Usando plt.bar()
'''

plt.bar(country,population)
plt.show
```

```python
'''
Podemos modificar el grosor de las barras con width= n
Modificamos los colores con color=[colorbarra1, colorbarra2....]
'''

plt.bar(country,population,width=0.2,color=['red','green','orange','white','gray'])
plt.show
```

```python
'''
Podemos modificar el grosor de las barras con width= n
Modificamos los colores con color=[colorbarra1, colorbarra2....]
Podemos hacer que los colores se alternen al declarar 2 colores
'''

plt.bar(country,population,width=0.2,color=['red','green'])
plt.show
```

```python
'''
Si queremos cambiar uno de los nombres de nuestras variables categoricas
directo desde la tabla:
'''

plt.bar(country,population,width=0.2,color=['red','green'])

'''
El arange solo nos ayudara a hacer las posiciones de cada barra, creando un arr
del 0 al 4. Como una especia de indice

Seguido definimos los nombres
'''
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'))

plt.show
```

```python
'''
Si quisieramos que las barras salieran de forma horizontal usamos
plt.barh
'''

plt.barh(country,population,height=0.2,color=['red','green'])
```

```python
# Definimos variables categoricas

country = ['VENEZUELA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [28200000,125700000,126700000,51520000,83200000]

pop_median = np.average(population)

plt.title('Countries Population')

plt.bar(country,population,width=0.5,color='gray')
plt.xticks(np.arange(5),(['VENEZUELA','JAPAN  ','MEXICO','COLOMBIA','GERMANY']),rotation=45)

plt.axhline(y = pop_median, color = 'purple', linestyle = ':', label='Population Median')

plt.show()
```
