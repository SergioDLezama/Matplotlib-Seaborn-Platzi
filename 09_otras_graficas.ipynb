```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('dark_background')
```

```python
# Definimos un array aleatorio
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data)

plt.show
```

```python
'''
Propiedades de los histogramas
bins= n Numero de barras donde llevaran la distribucion de datos
'''
plt.hist(data,bins=50)

plt.show
```

```python
'''
Propiedades de los histogramas
bins= n Numero de barras donde llevaran la distribucion de datos
width= n Grueso de las barras
'''
plt.hist(data,bins=25)

plt.show
```

```python
'''
Propiedades de los histogramas
histtype= 'tipo' Podemos cambiar el tipo de histograma
'''
plt.hist(data,bins=25,histtype='step')

plt.show
```

## Boxplot

```python
plt.boxplot(data)

plt.show()
```

```python
'''
Propiedades de boxplot
vert=False Para tener el boxplot horizontal
patch_artist=True Para rellenar el box
notch=True Nos ayuda a visualizar mejor la mediana y ver los datos concentrados
'''

plt.boxplot(data,vert=False,patch_artist=True,notch=True)

plt.show()
```

```python
'''
Propiedades de boxplot
Si agregamos un dato experior fuera del rango:
Maneja como un outlayer el dato exterior
'''

data = np.append(data,200)

plt.boxplot(data,vert=False,patch_artist=True,notch=True)

plt.show()
```

```python
'''
Propiedades de boxplot
Para no ver los datos exteriores usamos
showfliers=False
'''

data = np.append(data,200)

plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)

plt.show()
```

## Scatter

```python
# Definimos ciertas variables
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)

plt.show()
```

```python
'''
Propiedades de Scatter
s= Nnos ayuda a graficar una tercera variable
'''

plt.scatter(x,y,s=area)

plt.show()
```

```python
'''
Propiedades de Scatter
s= Nnos ayuda a graficar una tercera variable
c= Para definir colores
'''

plt.scatter(x,y,s=area,c=colors)

plt.show()
```

```python
'''
Propiedades de Scatter
s= Nnos ayuda a graficar una tercera variable
c= Para definir colores
marker= 'tipo' Para cambiar el tipo de marcador que queremos
'''

plt.scatter(x,y,s=area,c=colors,marker='x')

plt.show()
```

```python
'''
Propiedades de Scatter
s= Nnos ayuda a graficar una tercera variable
c= Para definir colores
marker= 'tipo' Para cambiar el tipo de marcador que queremos
alpha= n Para definir la transparencia de los marker
'''

plt.scatter(x,y,s=area,c=colors,marker='s',alpha=0.5)

plt.show()
```
