```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
'''Grafico de barras'''

sns.barplot(x=['A','B','C'],y=[1,3,2])

plt.show()
```

```python
'''
Comando set
style= 'Estilo' Cambia el estilo de el plot
palette= 'Paleta predefinida' Cambia los colores
font= 'Fuente' Cambia el estilo de letra
font_scale= n Tamano de la letra
'''

sns.set(style='darkgrid',palette='dark',font='Verdana',font_scale=2)

sns.barplot(x=['A','B','C'],y=[1,3,2])

plt.show()
```
