```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
# Importamos dataset de seaborn
DATA = sns.load_dataset('tips')

DATA.head()
```

## Distribuciones

```python
'''Histplot'''

sns.histplot(data=DATA,
             x= 'tip',
             bins=15,
             )

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
'''


sns.histplot(data=DATA,
             x= 'tip',
             bins=15,
             cumulative=True)

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False)

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'frequency' Para que el histograma no trabaje con el conteo sino la frequencia
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'frequency')

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'probability' Para que el histograma no trabaje con el conteo sino la probabilidad
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'probability')

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'percent' Para que el histograma no trabaje con el conteo sino el porcentaje
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'percent')

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'density' Para que el histograma no trabaje con el conteo sino la densidad
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'density')

plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'count' Para que el histograma trabaje con el conteo
multiple= '' La forma en la que queremos presentar los datos. 
             Layer en este ejemplo. Que hace el efecto en capas
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'count',
             multiple= 'layer')
plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'count' Para que el histograma trabaje con el conteo
multiple= '' La forma en la que queremos presentar los datos. 
             stack en este ejemplo. Que monta los datos uno encima del otro
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'count',
             multiple= 'stack')
plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'count' Para que el histograma trabaje con el conteo
multiple= '' La forma en la que queremos presentar los datos. 
             fill en este ejemplo. Que llena la tabla
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'count',
             multiple= 'fill')
plt.show()
```

```python
'''
Histplot

cumulative= True Para demostrar el efecto acomulativo
hue= '' Valor diferenciador
stat= 'count' Para que el histograma trabaje con el conteo
multiple= '' La forma en la que queremos presentar los datos. 
             dodge en este ejemplo. Que pone los datos uno al lado del otro
'''


sns.histplot(data=DATA,
             x= 'tip',
             hue='sex',
             bins=15,
             cumulative=False,
             stat= 'count',
             multiple= 'dodge')
plt.show()
```

```python
'''Diagrama de densidad'''
sns.kdeplot(data=DATA,
            x= 'tip',
            hue= 'sex')
        
plt.show()
```

```python
'''
Diagrama de densidad

cumulative= True Para ver los datos de corma acumulativa. Requiere scipy que no logro importar
'''
sns.kdeplot(data=DATA,
            x= 'tip',
            hue= 'sex',
            cumulative=True)
        
plt.show()
```

```python
'''
Diagrama de densidad

cumulative= True Para ver los datos de corma acumulativa. Requiere scipy que no logro importar
shade=True Llena el espacio representando lo lleno. Esta funcion esta descontinuada. Se debe usar fill=True
'''
sns.kdeplot(data=DATA,
            x= 'tip',
            hue= 'sex',
            cumulative=False,
            fill=True)
        
plt.show()
```

```python
'''
Diagrama de densidad

cumulative= True Para ver los datos de corma acumulativa. Requiere scipy que no logro importar
shade=True Llena el espacio representando lo lleno. Esta funcion esta descontinuada. Se debe usar fill=True
bw_adjust=n Ajusta la forma en la que vemos los datos

'''
sns.kdeplot(data=DATA,
            x= 'tip',
            hue= 'sex',
            cumulative=False,
            fill=True,
            bw_adjust=0.5)
        
plt.show()
```

```python
'''
Grafico escalonado
'''
sns.ecdfplot(data=DATA,x='tip',hue='sex')

plt.show()
```

```python
'''
Grafico escalonado
stat= '' Cambiamos la forma en evaluar los datos
'''
sns.ecdfplot(data=DATA,x='tip',hue='sex',stat='count')

plt.show()
```

```python
'''
Grafico de distribucion
En el curso se usa distplot() Pero esta descontinuado
Lo correcto es usar displot() en su lugar o histplot()

'''
sns.displot(data=DATA,
             x='tip',
             hue='sex')

plt.show( )
```

```python
'''
Grafico de distribucion
En el curso se usa distplot() Pero esta descontinuado
Lo correcto es usar displot() en su lugar o histplot()

kind='tipo' Para cambiar el tipo de grafico que queremos usar. Este argumento no se usa, es parte de distplot que esta descontinuado
'''
sns.displot(data=DATA,
             x='tip',
             hue='sex',
             kind='kde')

plt.show( )
```

```python
DATA['weekend'] = (DATA['day'] == 'Sun') | (DATA['day'] == 'Sat')
sns.displot(data=DATA,x='tip',hue='weekend',  kde=True, multiple='dodge')
plt.show()
```
