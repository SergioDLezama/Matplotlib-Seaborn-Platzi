```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargamos el data de tips
DATA = sns.load_dataset('tips')
DATA.head()
```

```python
'''
Countplot
Grafica la cuenta de la variable categorica, en este caso day,
Cuantos registros tengo de estos dias en la variable categorica
Va contar cuantas filas hay para los datos en el dia
'''

sns.countplot(data=DATA,x='day',hue='sex')
plt.show()
```

```python
'''
Countplot
Igualmente podemos graficarlo en el eje Y
'''

sns.countplot(data=DATA,y='day',hue='sex')
plt.show()
```

```python
'''
StripPlot
Crea un grafico de dispersion
'''

sns.stripplot(data=DATA,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
'''
StripPlot
Crea un grafico de dispersion
dodge= True Para separar los datos
'''

sns.stripplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
'''
StripPlot
Podemos aplicar los atributos de pyplot
'''
plt.figure(figsize=(6,6))


sns.stripplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
'''
Swarmplot
Similar a stripplot
A diferencia que tendremos un eje horizontal donde mas se concentran los datos
'''
plt.figure(figsize=(6,6))


sns.swarmplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
'''
Boxplot
Podemos ver la distribucion estadistica
'''
plt.figure(figsize=(6,6))


sns.boxplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
'''
Boxplot
Igual podemos graficar variables numericas, no solo categoricas
'''
plt.figure(figsize=(6,6))


sns.boxplot(data=DATA,y='total_bill')
plt.show()
```

```python
'''
Podemos apoyarnos entre varios graficos para visualizar mejor
'''
plt.figure(figsize=(6,6))


sns.boxplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True,color='0',marker='<')
plt.show()
```

```python
'''
Violinplot
Similar al boxplot pero sin percentiles
Demuestra como se concentran los datos
'''
plt.figure(figsize=(6,6))


sns.violinplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True,palette='pastel')
plt.show()
```

```python
'''
Violinplot
Similar al boxplot pero sin percentiles
Demuestra como se concentran los datos
'''
plt.figure(figsize=(6,6))


sns.violinplot(data=DATA,y='total_bill')
plt.show()
```

```python
'''
Violinplot
Similar al boxplot pero sin percentiles
Demuestra como se concentran los datos
split=True Para que los graficos se combinen
'''
plt.figure(figsize=(6,6))


sns.violinplot(data=DATA,x='day',y='total_bill',hue='sex',dodge=True,palette='pastel',split=True)
plt.show()
```

```python
'''
Catplot
Grafica variables categoricas
'''

sns.catplot(data=DATA,x='day',y='total_bill',hue='sex') # Nos grafica automaticamente un stripplot en este caso
plt.show()
```

```python
'''
Catplot
Grafica variables categoricas
kind='tipo'
col='Dato' Tercera varialble que nos ayuda a crear columnas
'''

sns.catplot(data=DATA,x='day',y='total_bill',hue='sex',kind='box',col='time')
plt.show()
```
