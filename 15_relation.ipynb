```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA = sns.load_dataset('tips')
DATA.head()
```

```python
'''
Scatterplot
'''
sns.scatterplot(data=DATA,x='total_bill',y='tip')

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
'''
sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day')

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
'''
sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day',style='time')

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
'''
sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size')

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
'''
sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size')

plt.legend(loc='center',bbox_to_anchor=(1.1,0.55))

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
'''
# Podemos definir los marker:
markers = {'Lunch':'D', 'Dinner':'s'}

sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)

plt.legend(loc='center',bbox_to_anchor=(1.1,0.55))

plt.show()
```

```python
'''
Scatterplot
hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
plt.figure(figsize=(x,y)) Para modificar el tamano de el plano
'''

plt.figure(figsize=(8,8))

# Podemos definir los marker:
markers = {'Lunch':'D', 'Dinner':'s'}

sns.scatterplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size',markers=markers)

plt.legend(loc='center',bbox_to_anchor=(1.1,0.5))

plt.show()
```

```python
'''Lineplot'''

sns.lineplot(data=DATA,x='total_bill',y='tip')

plt.show()
```

```python
'''
Lineplot
Recomendado para usar con pocas variables

hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
plt.figure(figsize=(x,y)) Para modificar el tamano de el plano
'''
plt.figure(figsize=(15,6))

sns.lineplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.05,0.5))

plt.show()
```

```python
'''
Relplot

kind='tipo' Modifica el tipo de plot

hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
plt.figure(figsize=(x,y)) Para modificar el tamano de el plano
'''
plt.figure(figsize=(15,6))

sns.relplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size')

plt.show()
```

```python
'''
Relplot

kind='tipo' Modifica el tipo de plot

hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
plt.figure(figsize=(x,y)) Para modificar el tamano de el plano
'''
plt.figure(figsize=(15,6))

sns.relplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size',kind='line')

plt.show()
```

```python
'''
Relplot

kind='tipo' Modifica el tipo de plot

hue='variable' Colorea por caracteristica a cada elemento
style='variable' Estiliza de diferente forma a los elementos segun la variable
size='Variable' Cambia de tamano cada elemento segun el valor de la variable
plt.legend(loc=) Para modificar la posicion de la leyenda
markers= Diccionario de los marcadores definidos
plt.figure(figsize=(x,y)) Para modificar el tamano de el plano
col='variable' Divide en columnas el plot
'''
plt.figure(figsize=(15,6))

sns.relplot(data=DATA,x='total_bill',y='tip',hue='day',style='time',size='size',col='time')

plt.show()
```
