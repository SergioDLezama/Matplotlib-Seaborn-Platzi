```python
import seaborn as sns
import matplotlib.pyplot as plt
DATA = sns.load_dataset('tips')
DATA.head()
```

```python
'''
Llevamos la data a una estructura matricial con
.corr(numeric_only=True)
Donde tenemos la correlacion entre las variables
Va de -1 a 1
Donde -1 es una relacion totalmente indirecta
Donde 0 No hay correlacion
Donde 1 Es una relacion directa
Todo lo que este en el medio representa a que tipo de relacion tenemos
'''

DATA.corr(numeric_only=True)
```

```python
'''Heatmap'''
sns.heatmap(DATA.corr(numeric_only=True))

plt.show()
```

```python
'''
Heatmap
annot=True Para tener los valores numericos tambien
'''
sns.heatmap(DATA.corr(numeric_only=True),annot=True)

plt.show()
```

```python
'''
Heatmap
annot=True Para tener los valores numericos tambien
cmap='paleta' Para cambiar la paleta de colores
linewidth=n Separa los cuadrados para mejor visualizacion
linecolor='Color' Color de la linea
'''
sns.heatmap(DATA.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=5,linecolor='black')

plt.show()
```

```python
'''
Heatmap
annot=True Para tener los valores numericos tambien
cmap='paleta' Para cambiar la paleta de colores
linewidth=n Separa los cuadrados para mejor visualizacion
linecolor='Color' Color de la linea
vmin=n Valor minimo
vmax=n Valor maximo
cbar=False Quita la linea leyenda
'''
sns.heatmap(DATA.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            linewidths=5,
            linecolor='black',
            vmin=1,
            vmax=1,
            cbar=False)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cond_dinner =DATA['time'] == 'Dinner' 
tips_dinner = DATA[cond_dinner] 
cond_lunch =(DATA['time'] =='Lunch')
tips_lunch = DATA[cond_lunch] 

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))
sns.heatmap(tips_dinner.corr(numeric_only=True), ax=ax1 ,annot=True,cmap='coolwarm',linewidths=5,linecolor='black')
sns.heatmap(tips_lunch.corr(numeric_only=True), ax=ax2,annot=True,cmap='coolwarm',linewidths=5,linecolor='black')
ax1.set_title('Dinner Correlation')
ax2.set_title('Lunch Correlation')
plt.show()
```
